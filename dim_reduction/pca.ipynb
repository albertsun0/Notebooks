```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
print(random.random())
```

```python
x = []
y = []
for i in range(100):
    for j in range(100):
        if random.random()/2 + 1/(abs(i - j) + 1)  > 0.5:
            x.append(i)
            y.append(j)
```

```python
plt.scatter(x,y)
```

```python
combined = [[a,b] for a,b in zip(x,y)]
```

```python
points = np.array(combined)
```

```python
standardized_points = (points - np.mean(points)) / np.std(points)
```
