# Lorenz System

Chaotic turbulent atmospheric convection


$\dot{x} = \sigma(y-x)$

$\dot{y} = x(p-z)$

$\dot{z} = xy-\beta z$

$B = [\sigma, p, \beta]^T$

$X = [x,y,z]$


```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# x[0] = x
# x[1] = y
# x[2] = z

# beta[0] = sigma
# beta[1] = rho
# beta[2] = beta

def lorenz(x,t,beta):
    dx = [
        beta[0] * (x[1] - x[0]),
        x[0] * (beta[1] - x[2]),
        x[0] * x[1] - beta[2] * x[2]
    ]
    return dx
```

```python
beta = [10,28,8/3]
x0 = [0, 1, 20]
dt = 0.001
t = np.linspace(0,50,5000)
```

```python
result = odeint(lorenz,x0,t,args=(beta,))
```

```python
result
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

X,Y,Z = result.T

ax.scatter(X,Y,Z, s=0.2)

plt.show()
```
